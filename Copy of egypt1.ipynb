```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('rice1.csv', sep = ',')
```

```python
data
```

```python
df = data.drop(["Element"], axis = 1)
```

```python
df
```

```python
Y = df.iloc[3].values
X=list(range(1961,2015))
j=[i.replace(",","") for i in Y]
k=[float(i) for i in j]
```

```python
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(X,k, color = 'red')
plt.title('Production vs Year')
plt.xlabel('Year')
plt.ylabel('Production')
plt.show()
```

```python
Y1= df.iloc[1].values
j1=[i.replace(",","") for i in Y1]
k1=[float(i) for i in j1]
```

```python
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(X,k1, color = 'red')
plt.title('Area vs Year')
plt.xlabel('Year')
plt.ylabel('Area')
plt.show()
```
